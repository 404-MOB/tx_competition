### 说明
* 这个文件主要是将0中得到的data.csv数据进行merge,因为我当时用的是train.csv,这样不会占用过多内存,所以得到1:1条目以后再进行连接
* 输入文件目录

data/adFeature.csv       #源数据文件

data/userFeature/userFeature_[1..111].csv   #切分出的111份userFeature文件

data/data.csv     # 0导出的label1:1文件

* 输出文件目录

data/01data/data.csv        #转换以后得到的完整数据集(label 1:1),数据条数4364806

```python
import pandas as pd
import numpy as np
import gc
import time
```

```python
def getData(path,ad_feature_name,user_feature_name,train_name,data_name):
    ad_feature = pd.read_csv(path+'/'+ad_feature_name+'.csv')
    user_feature = pd.read_csv(path+'/userFeature/'+user_feature_name+'.csv')
    train = pd.read_csv(path+'/'+train_name+'.csv')
    user_feature = user_feature.fillna(0)
    
    data = pd.merge(train,ad_feature,on='aid',how='left')
    data = pd.merge(data,user_feature,on='uid',how='left')
    data = (data[data.house >= -1])
    
    n = len(data)
    data.to_csv(path+'/01data/'+data_name+'.csv',index=False)
    
    del ad_feature
    del user_feature
    del train
    del data
    gc.collect()

    return n
```

```python
path = 'data'
ad_feature_name = 'adFeature'
user_feature_name = 'userFeature'
train_name = 'data'
data_name = 'data'
```

```python
def get_one_train_full_data(path,ad_feature_name,user_feature_name,train_name,data_name):
    count = 111
    num = 0;
    for i in range(1,count+1):
#         time1 = time.time()
        num += getData(path,ad_feature_name,user_feature_name+'_'+str(i),train_name,data_name+'_'+str(i))
        print(num)
#         if i == 1:
#             data = temp
#         else:
#             data = pd.concat([data,temp],ignore_index=True)
#         del temp
#         time2 = time.time()
#         print('%s merge %s_%d use time:%d\'s'%(train_name,user_feature_name,i,(time2-time1)))
#     print(len(data))
    print(num)
#     data.to_csv(path+'/01data/'+data_name+'.csv',index=False)
#     del data
#     gc.collect()
    #     time.sleep(10)
```

```python
s_1 = time.time()
get_one_train_full_data(path,ad_feature_name,user_feature_name,train_name,data_name)
s_2 = time.time()
print('%s merge %s totally use time:%d\'s'%(train_name,user_feature_name,(s_2-s_1)))
```

```python
def combineFile(filename, num):
    for i in range(1,num+1):
        try:
            load_filename = 'data/01data/data_%d.csv'%i
            fin = open(load_filename,'r')
            fout = open(filename,'a')
            head = fin.readline()
            if i == 1:
                fout.writelines([head])
            buf = []
            for line in fin:
                buf.append(line)
            fout.writelines(buf)
        finally:
            fin.close()
            fout.close()
```

```python
combineFile('data/01data/data.csv',111)
```
