```python
import pandas as pd
import numpy as np
import gc
import time
```

```python
def getData(path,ad_feature_name,user_feature_name,test_name):
    ad_feature = pd.read_csv(path+'/'+ad_feature_name+'.csv')
    user_feature = pd.read_csv(path+'/userFeature/'+user_feature_name+'.csv')
    test = pd.read_csv(path+'/test/'+test_name+'.csv')
    user_feature = user_feature.fillna(0)
    
    data = pd.merge(test,ad_feature,on='aid',how='left')
    data = pd.merge(data,user_feature,on='uid',how='left')
    data = (data[data.house >= -1])
    
    del ad_feature
    del user_feature
    del test
    gc.collect()
#     data.to_csv(path+'/data/'+data_name+'.csv',index=False)
    return data
```

```python

path = 'data'
ad_feature_name = 'adFeature'
user_feature_name = 'userFeature'
train_name = 'train'
test_name = 'test1'
data_name = 'test'
```

```python
def get_one_train_full_data(path,ad_feature_name,user_feature_name,test_name,data_name):
    count = 111
    for i in range(1,count+1):
#         time1 = time.time()
        temp = getData(path,ad_feature_name,user_feature_name+'_'+str(i),test_name)
        if i == 1:
            data = temp
        else:
            data = pd.concat([data,temp])
        del temp
#         time2 = time.time()
#         print('%s merge %s_%d use time:%d\'s'%(train_name,user_feature_name,i,(time2-time1)))
#     print(len(data))
    print(len(data))
    data.to_csv(path+'/data/'+data_name+'.csv',index=False)
    del data
    gc.collect()
```

```python
s_1 = time.time()
for i in range(1,119):
    start_time = time.time()
    get_one_train_full_data(path,ad_feature_name,user_feature_name,test_name+'_'+str(i),data_name+'_'+str(i))
    end_time = time.time()
    print('%s_%d merge %s use time:%d\'s'%(test_name,i,user_feature_name,(start_time-end_time)))
s_2 = time.time()
print('%s merge %s totally use time:%d\'s'%(test_name,user_feature_name,(s_1-s_2)))
```
