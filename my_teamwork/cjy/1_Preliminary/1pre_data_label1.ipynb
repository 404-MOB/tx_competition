```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
data = pd.read_csv("../data/data_label1.csv")
```

```python
int_feature=['aid','advertiserId','campaignId','creativeId','creativeSize','adCategoryId','productId','productType']
for feature in int_feature:
    features = data[feature].astype(str)
    plat = features.value_counts()

    plt.figure(figsize=(12,4))
    sns.barplot(plat.index, plat.values)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel(feature + 'count', fontsize=12)
```
