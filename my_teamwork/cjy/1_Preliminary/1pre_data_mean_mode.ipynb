```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
#data0 = pd.read_csv("../data/data1.csv")
data0 = pd.read_csv("../data/data.csv")
```

```python
data0.describe()                    #各项统计（仅对int型）
```

```python
data0.info()                        #看总数及数值类型
```

```python
int_feature=['aid','uid','advertiserId','campaignId','creativeId','creativeSize','adCategoryId','productId','productType','age',
         'gender','marriageStatus','education','consumptionAbility','LBS','ct','os','carrier','house']
```

```python
for feature in int_feature:
    # feature~label
    fig, (axis1,axis2,axis3) = plt.subplots(1,3,figsize=(15,5))
    #subplot1
    sns.countplot(x=feature, data=data0, ax=axis1)
    #subplot2
    sns.countplot(x='label', hue=feature, data=data0, order=[0,1,-1], ax=axis2)
    #subplot3
    embark_perc = data0[[feature, "label"]].groupby([feature],as_index=False).mean()
    sns.barplot(x=feature, y='label', data=data0,ax=axis3)
```
